## Exemples de code de Chat-GPT

```python
# Example DataFrame
df = pd.DataFrame({'Text1': ['This is a sentence.', 'Another example.', 'Text data.', 'Some text.'],
                   'Text2': ['Hello world!', 'Python programming.', 'Data analysis.', 'Machine learning.']})

# Columns to add noise
columns_with_noise = ['Text1', 'Text2']

# Add noise to selected columns
for column in columns_with_noise:
    # Iterate over each row in the column
    for i in range(len(df)):
        # Split the text into individual words
        words = df.loc[i, column].split()
        
        # Add noise to each word
        noisy_words = []
        for word in words:
            # Randomly choose a noise operation (e.g., adding random word or shuffling characters)
            noise_operation = random.choice(['add_word', 'shuffle_chars'])
            
            # Apply the noise operation
            if noise_operation == 'add_word':
                # Add a random word from the WordNet corpus
                synsets = wordnet.synsets(word)
                if synsets:
                    random_synset = random.choice(synsets)
                    random_word = random_synset.lemmas()[0].name()
                    noisy_words.append(random_word.replace('_', ' '))
            elif noise_operation == 'shuffle_chars':
                # Shuffle the characters within the word
                shuffled_word = ''.join(random.sample(word, len(word)))
                noisy_words.append(shuffled_word)
        
        # Join the noisy words back into a sentence
        noisy_text = ' '.join(noisy_words)
        
        # Update the DataFrame with the noisy text
        df.loc[i, column] = noisy_text

# Display the updated DataFrame
print(df)
```

```python
# Example DataFrame
df = pd.DataFrame({'Text1': ['This is a sentence.', 'Another example.', 'Text data.', 'Some text.'],
                   'Text2': ['Hello world!', 'Python programming.', 'Data analysis.', 'Machine learning.']})

# Columns to add noise
columns_with_noise = ['Text1', 'Text2']

# Function to introduce spelling mistakes in a word
def introduce_spelling_mistake(word):
    # Randomly choose a mistake operation (e.g., replacing character, adding or removing character)
    mistake_operation = random.choice(['replace', 'add', 'remove'])

    if mistake_operation == 'replace':
        # Replace a character with a similar-looking character
        letters = list(word)
        index = random.randint(0, len(letters) - 1)
        letters[index] = random.choice(list(letters[index]))
        return ''.join(letters)
    elif mistake_operation == 'add':
        # Add a random character after a randomly chosen position
        letters = list(word)
        index = random.randint(0, len(letters))
        letters.insert(index, random.choice(list('abcdefghijklmnopqrstuvwxyz')))
        return ''.join(letters)
    elif mistake_operation == 'remove':
        # Remove a character at a randomly chosen position
        letters = list(word)
        index = random.randint(0, len(letters) - 1)
        del letters[index]
        return ''.join(letters)
    else:
        return word

# Add noise to selected columns
for column in columns_with_noise:
    # Iterate over each row in the column
    for i in range(len(df)):
        # Split the text into individual words
        words = df.loc[i, column].split()
        
        # Add noise to each word
        noisy_words = []
        for word in words:
            # Apply spelling mistake only if the word has more than two characters
            if len(word) > 2:
                # Randomly choose whether to introduce a spelling mistake or not
                introduce_mistake = random.random() < 0.5  # Adjust the probability as per your requirement
                
                if introduce_mistake:
                    # Introduce a spelling mistake
                    noisy_word = introduce_spelling_mistake(word)
                    noisy_words.append(noisy_word)
                else:
                    noisy_words.append(word)
            else:
                noisy_words.append(word)
        
        # Join the noisy words back into a sentence
        noisy_text = ' '.join(noisy_words)
        
        # Update the DataFrame with the noisy text
        df.loc[i, column] = noisy_text

# Display the updated DataFrame
print(df)
```

```python
%%time 

df = mss.copy(deep=True)

# Columns to add noise
columns_with_noise = ['repository_key (Organisation) Record Title']

# Function to introduce spelling mistakes in a word
def introduce_spelling_mistake(word):
    # Randomly choose a mistake operation (e.g., replacing character, adding or removing character)
    mistake_operation = random.choice(['replace', 'add', 'remove'])

    if mistake_operation == 'replace':
        # Replace a character with a similar-looking character
        letters = list(word)
        index = random.randint(0, len(letters) - 1)
        letters[index] = random.choice(list(letters[index]))
        return ''.join(letters)
    elif mistake_operation == 'add':
        # Add a random character after a randomly chosen position
        letters = list(word)
        index = random.randint(0, len(letters))
        letters.insert(index, random.choice(list('abcdefghijklmnopqrstuvwxyz')))
        return ''.join(letters)
    elif mistake_operation == 'remove':
        # Remove a character at a randomly chosen position
        letters = list(word)
        index = random.randint(0, len(letters) - 1)
        del letters[index]
        return ''.join(letters)
    else:
        return word

# Add noise to selected columns
for column in columns_with_noise:
    # Iterate over each row in the column
    for i in range(len(df)):
        # Split the text into individual words
        words = df.loc[i, column].replace(' (Organisation)', '').split()
        
        # Add noise to each word
        noisy_words = []
        for word in words:
            # Apply spelling mistake only if the word has more than two characters
            if len(word) > 3:
                # Randomly choose whether to introduce a spelling mistake or not
                introduce_mistake = random.random() < 0.4  # Adjust the probability as per your requirement
                
                if introduce_mistake:
                    # Introduce a spelling mistake
                    word = introduce_spelling_mistake(word)
            
            noisy_words.append(word)
        
        # Join the noisy words back into a sentence
        noisy_text = ' '.join(noisy_words)
        
        # Update the DataFrame with the noisy text
        df.loc[i, column] = noisy_text

# Display the updated DataFrame
print(df)
```
